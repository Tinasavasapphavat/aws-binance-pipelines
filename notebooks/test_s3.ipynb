```python
import boto3
import pandas as pd
import io
from clickhouse_driver import Client
from typing import Optional
from tqdm import tqdm
```

```python

s3 = boto3.client('s3')
bucket = 'databricks-binance-pipeline'

# List all parquet files
paginator = s3.get_paginator('list_objects_v2')
files = []
for page in paginator.paginate(Bucket=bucket):
    for obj in page.get('Contents', []):
        if obj['Key'].endswith('.parquet'):
            files.append(obj['Key'])

print(f'Total files found: {len(files)}')
# for f in files:
#     print(f)

# Load all into one dataframe
dfs = []
btc = []
eth = []
for file_key in tqdm(files):
    print(file_key)
    if "ETH" in file_key:
        obj = s3.get_object(Bucket=bucket, Key=file_key)
        df = pd.read_parquet(io.BytesIO(obj['Body'].read()))
        eth.append(df)
    if "BTC" in file_key:
        obj = s3.get_object(Bucket=bucket, Key=file_key)
        df = pd.read_parquet(io.BytesIO(obj['Body'].read()))
        btc.append(df)

btc_df = pd.concat(btc, ignore_index=True)
eth_df = pd.concat(eth, ignore_index=True)
# combined_df = pd.concat(dfs, ignore_index=True)
# print(f'Total rows: {len(combined_df)}')
# print(f'Columns: {combined_df.columns.tolist()}')
# print(combined_df.head(3))
```

```python


def prepare_df(df):
    df = df.copy()

    df['datetime']       = pd.to_datetime(df['datetime'], errors='coerce')
    df['is_maker']       = df['is_maker'].astype(bool)
    df['best_match']     = df['best_match'].astype(bool)
    df['trade_id']       = pd.to_numeric(df['trade_id'], errors='coerce')
    df['first_trade_id'] = pd.to_numeric(df['first_trade_id'], errors='coerce')
    df['last_trade_id']  = pd.to_numeric(df['last_trade_id'], errors='coerce')
    df['timestamp_utc']  = pd.to_numeric(df['timestamp_utc'], errors='coerce')
    df['price']          = pd.to_numeric(df['price'], errors='coerce')
    df['quantity']       = pd.to_numeric(df['quantity'], errors='coerce')

    # drop bad rows (important!)
    df = df.dropna()

    return df

```
